```python
import numpy as np 
import matplotlib.pyplot as plt 

# random number generation 
# 2-D data 
num_samples=500
X = np.random.randint(10,100,size=(num_samples))
Y = np.random.randint(10,100,size=(num_samples))

# lest visualise the 2D data 
plt.figure(1)
plt.plot(X,Y,'ro')
# markers - o-circle, d-diamond, *-star,s - square

plt.title('Random Raw Data')
print(X.shape)
print(Y.shape)
```

```python
# 
mn = X.min()
mx = X.max()
midValue = int((mn+mx)/2)
print(midValue)
```

```python
# traverse each sample of X axis and shift the data point as per its value either in c1 or in c2 

plt.figure(2)
for i in range(num_samples):
    if(X[i]<=midValue):
        plt.plot(X[i],Y[i],'bd')
    elif(X[i]>midValue):
        plt.plot(X[i],Y[i],'go')

    
    
```

```python
plt.figure(2,figsize=(12,12))
plt.subplot(2,1,1)
plt.plot(X,Y,'ro')
plt.title('Raw Data')
plt.subplot(2,1,2)
for i in range(num_samples):
    if(X[i]<=midValue):
        plt.plot(X[i],Y[i],'bd')
    elif(X[i]>midValue):
        plt.plot(X[i],Y[i],'go')
plt.title('Clustered Data')
```

```python
plt.figure(2,figsize=(12,12))
plt.subplot(2,1,1)
plt.plot(X,Y,'ro')
plt.title('Raw Data')
plt.subplot(2,1,2)
for i in range(num_samples):
    if(X[i]<=midValue and Y[i]<=midValue):
        plt.plot(X[i],Y[i],'bd')
    elif(X[i]<=midValue and Y[i]>midValue):
        plt.plot(X[i],Y[i],'go')
    elif(X[i]>midValue and Y[i]<=midValue):
        plt.plot(X[i],Y[i],'ro')
    elif(X[i]>midValue and Y[i]>midValue):
        plt.plot(X[i],Y[i],'mo')
    
        
plt.title('Clustered Data')

```
